# Import the Data with traditional matlab

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis)
```

# 5.1.4 Annotate Charts

```python
#create the plot and add a label or the legend
plt.plot(x_axis, y_axis, label='Boston')

#Create labels for the x and y axis
plt.xlabel("Date")
plt.ylabel("Fare($)")

#set the y limit between 0 and 45
plt.ylim(0,45)

#Create a title
plt.title("PyBer Fare by Month")

#Add the legend
plt.legend()
```

```python
#change width, colour, add grid

#create the plot and add a label or the legend
plt.plot(x_axis, y_axis,marker ="*", color = "blue", linewidth=2, label='Boston')

#Create labels for the x and y axis
plt.xlabel("Date")
plt.ylabel("Fare($)")

#set the y limit between 0 and 45
plt.ylim(0,45)

#Create a title
plt.title("PyBer Fare by Month")

# ADD A GRID
plt.grid()

#Add the legend
plt.legend()
```

# Create graph with OOP and change formatting

```python
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, label='Boston', color="Green", marker=".", linewidth=2)

#add labels
ax.set_title("PyBer Fare by Month")
ax.set_ylabel("Fares($)")
ax.set_xlabel("Month")

#add gridlines
ax.grid()

#add legend
ax.legend()
```

# 5.1.5: Create Bar Charts using the MATLAB Approach

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.bar(x_axis, y_axis, color="green", label='Boston')

#Create labels
plt.xlabel("Date")
plt.ylabel("Fare($)")

#Create title
plt.title("PyBer Fare by month")

#add legent
plt.legend()
```

```python
#create horizontal chart
plt.barh(x_axis, y_axis, color="magenta", label='Boston')

#invert the y axis
plt.gca().invert_yaxis()

#invert the x axis
# plt.gca().invert_xaxis()

#Create labels
plt.xlabel("Date")
plt.ylabel("Fare($)")

#Create title
plt.title("PyBer Fare by month")

#add legent
plt.legend()
```

# 5.1.6 Create a bar chart with Object oriented approach

```python
#Create vertical bar graph with oom
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color="k")
```

```python
#Create vertical bar graph with oom
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")

ax.set_title("PyBer Fare by Month")
ax.set_ylabel("Month")
ax.set_xlabel("Fare($)")

ax.legend()
```

# 5.1.7: Creating Scatter plots and bubble charts

```python
#create a scatter plot with plt.plot
plt.plot(x_axis, y_axis, 'o')
```

```python
#create a scatter plot with plt.scatter()
plt.scatter(y_axis, x_axis, marker=".", s=100, c='red')

plt.xlim(0, 50)
plt.gca().invert_yaxis()
plt.xlabel("Fare($)")
plt.ylabel("Date")
```

```python
#Creating a bubble chart using matlab
```

```python
# plt.scatter(x_axis, y_axis, s=y_axis)

#make the points bigger by going through the y list and multiplying by a number
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# List comprehension
plt.scatter(x_axis, y_axis, s=[i*3 for i in y_axis])
```

```python
#5.1.7 scatter with object oriented interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis,s=[i*5 for i in y_axis], 
          c="skyblue", alpha=0.8, edgecolors='black', linewidth=2)
```

# 5.1.8 creating Pie Charts

```python
#matlap approoach
plt.subplots(figsize=(8,8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0.3, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.pie(y_axis, 
        explode=explode_values, 
        labels=x_axis, 
        colors = colors,
        autopct="%.1f%%")
plt.show()
```

```python
#Object Oriented Interface: Pie Charts
fig, ax = plt.subplots(figsize=(8,8))
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
ax.pie(y_axis,
       labels=x_axis,
       colors=colors,
      autopct="%.1f%%",
      explode=explode_values,
      startangle=90,
      shadow=True)

plt.gca().invert_xaxis()

plt.show()
```
